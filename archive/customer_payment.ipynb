```python
# customer payment 
# import libraries
import pandas as pd
from faker import Faker
import random
```

```python
# create faker object
f = Faker()

# set seed
random.seed(2)

# creating unique ids
def unique_ids(n):
    uid = [f.unique.random_number(digits = 3) for _ in range(n)]
    return uid
def input_data(x):
    customer_id = unique_ids(x)

    data = pd.DataFrame()
    for i in range(0, x):
        data.loc[i, "customer_id"] = customer_id[i]
        data.loc[i, "total_hours_day"] = random.choice([4, 8])
        data.loc[i, "total_hours_week"] = data.loc[i, "total_hours_day"] * 5
        data.loc[i, "total_hours_month"] = data.loc[i, "total_hours_week"] * 4
        data.loc[i, "monthly_payment"] = data.loc[i, "total_hours_month"] * 10
        data.loc[i, "total_payment_year"] = data.loc[i, "monthly_payment"] * 12
    return data
result = input_data(100)
result.to_csv("customer_payment.csv", index = False)
```
