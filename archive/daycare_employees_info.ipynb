```python
# Code to create an employee dataset
# import libraries
import pandas as pd
from faker import Faker
import random
```

```python
# create faker object
f = Faker()

# set seed
random.seed(2)

# mapping salary for specific position
salary_mapping = {
    "Manager": 40000,
    "Manager Assistant": 21000,
    "Teacher": 30000,
    "Teacher Assistant": 27000,
    "Cheff": 23000,
    "Maintenance": 20000,
    "Nurse": 35000,
}

# Create a dictionary of job title and number of employees in specified job title.
job_title_counts = {
    "Manager": 1,
    "Manager Assistant": 1,
    "Nurse": 1,
    "Cheff": 2,
    "Teacher": 9,
    "Teacher Assistant": 6,
    "Maintenance": 3
}

# calculate the total number of employees according to job title count
total_employees = sum(job_title_counts.values())


def input_employee(total_employees):
    employees_data = [] # list to store employee data
    employee_id = 22 # initialize employee id

    remaining_employees = total_employees

    # gen a while loop if there is > 0 employees
    while remaining_employees > 0:

        # choose a job title based on remaining count
        available_job_titles = [title for title, count in job_title_counts.items() if count > 0]
        job_title = random.choice(available_job_titles)

        # update the job title count and remaining employees
        job_title_counts[job_title] -= 1
        remaining_employees -= 1

        # map salary to get the salary based on the job title
        salary = salary_mapping[job_title]

        employee = {
            "employee_id": employee_id,
            "employee_name": f.name(),
            "birthdate": f.date_of_birth(minimum_age=18, maximum_age=70),
            "address": f.address(),
            "email": f.email(),
            "job_title": job_title,
            "salary_year": salary,
            "latitude": str(f.latitude()),
            "longitude": str(f.longitude())
        } 
        #append the each iteration in employees_data list
        employees_data.append(employee)
        employee_id +=1
    #create dataframe using the list
    data = pd.DataFrame(employees_data)
    return data

# generate data and save in the result
result = input_employee(total_employees)

#save the generated data to a csv file
result.to_csv("daycare_employee_info.csv", index = False)
```
